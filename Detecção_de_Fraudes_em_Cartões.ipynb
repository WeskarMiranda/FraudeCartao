```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
# Carregando o dataset
data = pd.read_csv('/content/creditcard - menor balanceado.csv')
```

```python
display(data.head())
```

```python
display(data.isnull().sum().to_frame("Valores Nulos"))
```

```python
print("DimensÃµes do dataset:", data.shape)
```

```python
# Separando as features e a variÃ¡vel alvo
X = data.drop('Class', axis=1)
y = data['Class']
print("DimensÃµes de X (features):", X.shape)
print("DimensÃµes de y (alvo):", y.shape)
```

```python
# Dividindo em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
print("DimensÃµes do conjunto de treino:", X_train.shape, y_train.shape)
print("DimensÃµes do conjunto de teste:", X_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# 1. SeleÃ§Ã£o de Atributos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
display(pd.DataFrame(X_train_scaled, columns=X.columns).head())
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
# Selecionando as k melhores features
k = 30
select_k_best = SelectKBest(f_classif, k=k)
X_train_kbest = select_k_best.fit_transform(X_train_scaled, y_train)
X_test_kbest = select_k_best.transform(X_test_scaled)
print("DimensÃµes apÃ³s SelectKBest (treino):", X_train_kbest.shape)
print("Exemplo de dados de treino apÃ³s SelectKBest:")
display(pd.DataFrame(X_train_kbest).head())
```

A utilizaÃ§Ã£o de mÃ©todos como SelectKBest reduziu as features para as mais informativas, diminuindo a complexidade e potencialmente melhorando a capacidade do modelo de identificar padrÃµes associados a fraudes. Os dados de treino apÃ³s essa seleÃ§Ã£o de atributos mostram que os valores foram normalizados e padronizados, facilitando o treinamento de modelos de ML e contribuindo para a precisÃ£o.

```python
from imblearn.over_sampling import SMOTE
```

```python
# 2. Balanceamento dos Dados
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_kbest, y_train)
print("DimensÃµes do conjunto balanceado de treino:", X_train_balanced.shape, y_train_balanced.shape)
print("DistribuiÃ§Ã£o das classes apÃ³s SMOTE:")
display(pd.DataFrame(y_train_balanced, columns=['Class']).value_counts().to_frame("Quantidade"))
```

O uso do SMOTE para balancear as classes ajustou o conjunto de dados para que houvesse uma distribuiÃ§Ã£o igual de transaÃ§Ãµes legÃ­timas e fraudulentas (887 de cada). Essa abordagem Ã© especialmente relevante, dado o desbalanceamento original, e foi aplicada para reduzir o viÃ©s do modelo para a classe majoritÃ¡ria, aumentando a sensibilidade em identificar fraudes.

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
```

```python
# 3. Tuning de HiperparÃ¢metros
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}
param_grid_svm = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf']
}

rf = RandomForestClassifier(random_state=42)
grid_rf = GridSearchCV(rf, param_grid_rf, scoring='f1', cv=5)
grid_rf.fit(X_train_balanced, y_train_balanced)
print("Melhores parÃ¢metros para Random Forest:", grid_rf.best_params_)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(probability=True, random_state=42)
grid_svm = GridSearchCV(svm, param_grid_svm, scoring='f1', cv=5)
grid_svm.fit(X_train_balanced, y_train_balanced)
print("Melhores parÃ¢metros para SVM:", grid_svm.best_params_)
```

```python
# Melhor modelo e parÃ¢metros de Random Forest e SVM
best_rf = grid_rf.best_estimator_
best_svm = grid_svm.best_estimator_
```

Os melhores parÃ¢metros encontrados para o Random Forest e o SVM, atravÃ©s de GridSearchCV ou RandomizedSearchCV, mostram a importÃ¢ncia de ajustar os hiperparÃ¢metros para maximizar o desempenho. A utilizaÃ§Ã£o de max_depth=None e min_samples_split=5 para Random Forest e C=10 para SVM, por exemplo, ajudam a modelar melhor os dados e reduzir overfitting.

```python
# 4. Ensemble de Modelos
voting_clf = VotingClassifier(estimators=[
    ('rf', best_rf),
    ('svm', best_svm),
    ('gb', GradientBoostingClassifier(random_state=42))
], voting='soft')
voting_clf.fit(X_train_balanced, y_train_balanced)
```

A combinaÃ§Ã£o de Random Forest, SVM e Gradient Boosting por meio de um Voting Classifier, utilizando a estratÃ©gia de votaÃ§Ã£o soft, aumentou a robustez do modelo e permitiu o aproveitamento dos pontos fortes de cada tÃ©cnica. Essa abordagem contribuiu para melhorar a precisÃ£o e a capacidade de generalizaÃ§Ã£o do modelo, um ponto essencial para sistemas de detecÃ§Ã£o de fraude.

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
```

```python
# 5. AvaliaÃ§Ã£o do Modelo
y_pred = voting_clf.predict(X_test_kbest)

# MÃ©tricas de avaliaÃ§Ã£o
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, output_dict=True)

# Exibindo os resultados como tabelas
print("\nResultados da AvaliaÃ§Ã£o:")
print("AcurÃ¡cia:", accuracy)
print("F1-Score (Macro):", f1)
print("Matriz de ConfusÃ£o:")
display(pd.DataFrame(conf_matrix, index=["Classe 0", "Classe 1"], columns=["Previsto 0", "Previsto 1"]))
print("RelatÃ³rio de ClassificaÃ§Ã£o:")
display(pd.DataFrame(class_report).transpose())
```

As mÃ©tricas de avaliaÃ§Ã£o indicam um bom desempenho geral:

AcurÃ¡cia: 96%, o que Ã© alto, mas dado o cenÃ¡rio desbalanceado, a acurÃ¡cia nÃ£o Ã© o foco principal.

F1-Score Macro: 92,6%, confirmando um bom equilÃ­brio entre precisÃ£o e recall para as classes.

Matriz de ConfusÃ£o: O modelo prevÃª com precisÃ£o a maioria das transaÃ§Ãµes, com 376 acertos e apenas 4 erros para a classe 0, e 131 acertos com 17 erros para a classe 1 (fraudulenta).

RelatÃ³rio de ClassificaÃ§Ã£o: A precisÃ£o e o recall estÃ£o bem distribuÃ­dos, com um F1-score de 97,2% para a classe legÃ­tima e 92,6% para a classe fraudulenta. O recall de 88,5% na classe fraudulenta indica que o modelo estÃ¡ conseguindo identificar a maioria das fraudes, ainda que haja um pequeno nÃºmero de falsos negativos.
